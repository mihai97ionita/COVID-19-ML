```python
from pandas import read_csv, DataFrame
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from matplotlib import pyplot
pyplot.rcParams["figure.figsize"] = (20,10)
```

```python
datasets = []

csv_testing_vaccine = '../../data_set2/data_vaccine_map/dataset_vaccine_and_testing.csv'
data = read_csv(csv_testing_vaccine, index_col=0, parse_dates=True, squeeze=True)
datasets.append(data)
```

```python
dataset_number = 0
country_code = 'RO'
dataset = datasets[dataset_number]
working_dataset_vaccine = dataset[dataset['ReportingCountry']==country_code].copy().drop('ReportingCountry', axis='columns')

working_dataset_vaccine["positivity_rate"].plot()
pyplot.legend(['positivity_rate'])
pyplot.show()

working_dataset_vaccine["TotalDosesSum"].plot()
pyplot.legend(['TotalDosesSum'])
pyplot.show()

print("dataset size -- {}".format(len(working_dataset_vaccine)))
```
