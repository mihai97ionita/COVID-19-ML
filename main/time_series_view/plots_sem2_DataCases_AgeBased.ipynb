```python
from pandas import read_csv, DataFrame
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from matplotlib import pyplot
pyplot.rcParams["figure.figsize"] = (20,10)
```

```python
csv = '../../data_set2/data_agebased_map/vaccine_ageBased_merged.csv'
data = read_csv(csv, index_col=0, parse_dates=True, squeeze=True)
```

```python
country_code = 'AT'

working_dataset = data[data['ReportingCountry']==country_code].copy().drop('ReportingCountry', axis='columns')
dataset_15_24 = working_dataset[working_dataset["TargetGroup"] == "15-24yr"].copy().drop('TargetGroup', axis='columns')
dataset_25_49 = working_dataset[working_dataset["TargetGroup"] == "25-49yr"].copy().drop('TargetGroup', axis='columns')
dataset_50_79 = working_dataset[working_dataset["TargetGroup"] == "50-79yr"].copy().drop('TargetGroup', axis='columns')
dataset_80plus = working_dataset[working_dataset["TargetGroup"] == "80+yr"].copy().drop('TargetGroup', axis='columns')

print("15-24yr")
dataset_15_24["TotalDosesSum"].plot()
pyplot.legend(['TotalDosesSum'])
pyplot.show()
dataset_15_24["new_cases"].plot()
pyplot.legend(['new_cases'])
pyplot.show()

print("25-49yr")
dataset_25_49["TotalDosesSum"].plot()
pyplot.legend(['TotalDosesSum'])
pyplot.show()
dataset_25_49["new_cases"].plot()
pyplot.legend(['new_cases'])
pyplot.show()

print("50-79yr")
dataset_50_79["TotalDosesSum"].plot()
pyplot.legend(['TotalDosesSum'])
pyplot.show()
dataset_50_79["new_cases"].plot()
pyplot.legend(['new_cases'])
pyplot.show()

print("80+yr")
dataset_80plus["TotalDosesSum"].plot()
pyplot.legend(['TotalDosesSum'])
pyplot.show()
dataset_80plus["new_cases"].plot()
pyplot.legend(['new_cases'])
pyplot.show()
```
